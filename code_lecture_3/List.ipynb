# List Operations

Using List comprehension Generate a list of squares of numbers from 1 to 10.

```python
square_list = [num**2 for num in range (1,11)]
print(square_list)
```

Using List Comprehension Create a list of even numbers from 1 to 20


```python
even_num= [num for num in range (1,21) if num%2==0]
print(even_num)
```

Find the minimum, maximum, length, and sum of a list.

```python
number=[num for num in range (0,101,10)]
print(number)
print(min(number))
print(max(number))
print(len(number))
print(sum(number))
```
