Loop through a Dictionary

```python
#loop through Dictionary

data ={'a':1,'b':2,'c':3, 'd':3}
#print data
for key,value in data.items():
  print(f"key: {key} and value: {value}")
```

Find and print all keys that have values greater than a given number.

```python
Number={'a':1,'b':2,'c':3}
#taking input
threshold = int(input('give me the number among 1 to 3: '))
#print data
for key,value in Number.items():
  if value>threshold:
      print(f"key: {key} and value: {value} is greater than threshold value {threshold}")
  else:
      print(f"key: {key} and value: {value} is not greater than threshold value {threshold}")
```

Count the occurrences of each value in a dictionary.

```python
# Initialize an empty dictionary to hold the counts
value_count = {}

# Loop through each value in the data dictionary
for value in data.values():
 # If the value is already in the value_count dictionary, increment its count
    if value in value_count:
        value_count[value] += 1
    # If the value is not in the dictionary, add it with a count of 1
    else:
        value_count[value] = 1

print(value_count)
```
