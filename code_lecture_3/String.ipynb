# String Slicing & Iteration

1.   Extract the first 5 and last 5 characters from a given string.
2.   Remove every alternate character from a string.
3. Extract the substring from index 2 to 7.
4. Iterate through a string and print each character.
5. Reverse a string using slicing.




Extract the first 5 and last 5 characters from a given string.

```python
string='We love Python'
#print string
first_five=string[:5]
last_five=string[:-5]

print(first_five)
print(last_five)
```

Remove every alternate character from a string.

```python
alternate_removed=string[::2]
print(alternate_removed)
```

Extract the substring from index 2 to 7.

```python
substring=string[2:8]
print(substring)
```

Iterate through a string and print each character.

```python
for character in string:
  print(character)
```

Reverse a string using slicing.

```python
reverse = string[::-1]
print(reverse)
```

Remove leading spaces using lstrip().



```python
string1=input('enter a space with leading and trailing spaces: ')
print(string1.lstrip())
print(string1.rstrip())
print(string1.lstrip().rstrip())
print(string1.strip())
```
